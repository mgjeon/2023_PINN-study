```python
from sunpy.map import Map
import math
import numpy as np
```

```python
radsindeg = np.pi/180.
```

```python
file_bz = '/userhome/jeon_mg/workspace/PINN_NF2/data/SHARP/NOAA12673/hmi.sharp_cea_720s.7115.20170906_083600_TAI.Br.fits'
```

```python
bz_map = Map(file_bz)
```

```python
header = bz_map.meta
```

```python
rsun_ref = header['rsun_ref']
dsun_obs = header['dsun_obs']
rsun_obs = header['rsun_obs']
cdelt1   = header['cdelt1']

bz                = bz_map.data
nx     = bz.shape[1]
ny     = bz.shape[0]
```

```python
cdelt1_arcsec = (math.atan((rsun_ref*cdelt1*radsindeg)/(dsun_obs)))*(1/radsindeg)*(3600.)
cdelt1_arcsec
```

```python
# (pix/arcsec)(arsec/meter)(meter/Mm)
# =(1/cdelt1_arcsec)(RSUN_OBS/RSUN_REF)(1000000)
# (pix/Mm)
unitconstant = (1/cdelt1_arcsec)*(rsun_obs/rsun_ref)*(10e6)
unitconstant
```

```python
# (Mm/pix)
k = 1/unitconstant
k
```
