```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime    
def get_free_energy(csv_path):
    csvtable = pd.read_csv(csv_path)
#     csvtable.head()

#     csvtable.keys()

    # csvtable[' Ef (10^33 erg)'].plot()

    x = csvtable['TIME (10^9 sec since 1-jan-1979 )']*1e9
    x = x.map(lambda x: datetime.datetime(1979, 1, 1) + datetime.timedelta(seconds=x))
    y = csvtable[' Ef (10^33 erg)']*10
    return x, y
```

```python
x, y = get_free_energy('/Users/mgjeon/Workspace/PINN_NF2/database/author/series_01.csv')
plt.plot(x,y, label='series_0.1')

x, y = get_free_energy('/Users/mgjeon/Workspace/PINN_NF2/database/author/series_1.csv')
plt.plot(x,y, label='series_1')

x, y = get_free_energy('/Users/mgjeon/Workspace/PINN_NF2/database/author/series_wb1.csv')
plt.plot(x,y, label='series_wb1')

x, y = get_free_energy('/Users/mgjeon/Workspace/PINN_NF2/database/author/series_wb2.csv')
plt.plot(x,y, label='series_wb2')

plt.ylabel('Free Energy (10^32 erg)')
plt.legend()
```
