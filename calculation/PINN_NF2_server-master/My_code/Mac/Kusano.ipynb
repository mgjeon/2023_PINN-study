```python
GD = Path('/Users/mgjeon/Library/CloudStorage/GoogleDrive-A/My Drive/Notes/Work/PINN/')
```

```python
from sunpy.map import Map
```

```python
hmi_p = GD / 'AR_377_series/hmi.sharp_cea_720s.377.20110213_120000_TAI.Bp.fits'
hmi_t = GD / 'AR_377_series/hmi.sharp_cea_720s.377.20110213_120000_TAI.Bt.fits'
hmi_r = GD / 'AR_377_series/hmi.sharp_cea_720s.377.20110213_120000_TAI.Br.fits'

Bp = Map(hmi_p).data
Bt = Map(hmi_t).data
Br = Map(hmi_r).data
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
vmin = -3000
vmax = 3000

fig, axes = plt.subplots(1,3, figsize=(15, 5))
im = axes[0].imshow(Bp, origin='lower', cmap='gray', vmin=vmin, vmax=vmax)
axes[0].set_title(f'{np.shape(Bp)} Bp [G]')
axes[1].imshow(-Bt, origin='lower', cmap='gray', vmin=vmin, vmax=vmax)
axes[1].set_title(f'{np.shape(Bt)} Bt [G]')
axes[2].imshow(Br, origin='lower', cmap='gray', vmin=vmin, vmax=vmax)
axes[2].set_title(f'{np.shape(Br)} Br [G]')

fig.subplots_adjust(bottom=0.2)
cbar_ax = fig.add_axes([0.36, 0.2, 0.3, 0.03])
fig.colorbar(im, cax=cbar_ax, orientation='horizontal')

fig.suptitle(t='NOAA 11158_20110213_120000', x=0.5, y=0.87)
fig.savefig('./Obs.png',dpi=300)
```
