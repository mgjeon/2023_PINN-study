```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
metric_path = './output/eval'

df = pd.read_csv(os.path.join(metric_path, 'metric.csv'))
```

```python
iter = np.array(df['iter'])
C_vec = np.array(df['C_vec'])
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.hlines(C_vec[0], xmin=iter[1], xmax=iter[-1], color='red', label='LL')
ax.plot(iter[1:], C_vec[1:], color='blue', label='PINN')
ax.legend(loc='center right')
ax.set_xlabel('Iteration', fontsize='medium')
ax.set_ylabel('C_vec', fontsize='medium')
fig.show()
```
